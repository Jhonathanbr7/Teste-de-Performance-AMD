```python
# Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV e armazenar os dados em um DataFrame
tabela = pd.read_csv("base.csv")

# Criar uma lista de valores percentuais a cada 20%
percentuais = [f"{i}% " for i in range(0, 101, 20)]

# Definir o número de ticks desejados no eixo x
num_ticks = len(percentuais)

# Criar o gráfico
plt.figure(figsize=(15, 5))  # Definir o tamanho da figura
plt.ylim(0, tabela.FPS.max()*1.20)  # Definir o limite do eixo y
plt.ylabel('FPS')  # Definir o rótulo do eixo y
plt.xlabel('Quantidade de Capturas')  # Definir o rótulo do eixo x

# Definir os locais dos ticks manualmente para corresponder aos rótulos de 20%
tick_locations = [i * (len(tabela) // (num_ticks - 1)) for i in range(num_ticks)]

# Definir os rótulos personalizados no eixo x
plt.xticks(tick_locations, percentuais)

# Plotar o gráfico com os dados de FPS
plt.title("FPS")
plt.plot(tabela.FPS)

# Exibir o gráfico
plt.show()
```

```python
# Importar as bibliotecas necessárias
import pandas as pd
import matplotlib.pyplot as plt

# Ler o arquivo CSV e armazenar os dados em um DataFrame
tabela = pd.read_csv("base.csv")
tabela = tabela.rename(columns={"UTIL. MEM. SISTEMA":"MEM"})
# Criar uma lista de valores percentuais a cada 20%
percentuais = [f"{i}% " for i in range(0, 101, 20)]

# Definir o número de ticks desejados no eixo x
num_ticks = len(percentuais)

# Criar o gráfico
plt.figure(figsize=(15, 5))  # Definir o tamanho da figura
plt.ylim(tabela.MEM.min()*0.50, tabela.MEM.max()*1.20)  # Definir o limite do eixo y
plt.ylabel('GB RAM')  # Definir o rótulo do eixo y
plt.xlabel('Quantidade de Capturas')  # Definir o rótulo do eixo x

# Definir os locais dos ticks manualmente para corresponder aos rótulos de 20%
tick_locations = [i * (len(tabela) // (num_ticks - 1)) for i in range(num_ticks)]

# Definir os rótulos personalizados no eixo x
plt.xticks(tick_locations, percentuais)

# Plotar o gráfico com os dados de Memória
plt.title("Utilização de Memória RAM")
plt.plot(tabela.MEM)

# Exibir o gráfico
plt.show()
```
